```python
import librosa
import numpy as np
import os
import soundfile
from IPython.display import Audio

from google.colab import drive
drive.mount('/content/drive')
```

```python
AUDIO_PATH = '/content/drive/MyDrive/cs224s_spring2022/data/harpervalleybank'
agent_audio_path = os.path.join(AUDIO_PATH, 'audio/agent')
caller_audio_path = os.path.join(AUDIO_PATH, 'audio/caller')
```

```python
agent_pp_audio_path = os.path.join(AUDIO_PATH, 'pp-audio/agent')
caller_pp_audio_path = os.path.join(AUDIO_PATH, 'pp-audio/caller')

#os.mkdir(agent_pp_audio_path)
#os.mkdir(caller_pp_audio_path)
```

```python
import preprocess
```

```python
librosa.__version__
```

```python
def preprocess_audio(src, dest):
  raw_audio, sr = librosa.load(src)
  pp_audio = preprocess.preprocess(raw_audio)
  print(f"Writing to {dest}")
  soundfile.write(dest, pp_audio, sr)
```

```python
audio_name = np.random.choice(os.listdir(agent_audio_path))
src_path = os.path.join(agent_audio_path, audio_name)
dest_path = os.path.join(agent_pp_audio_path, audio_name)
```

```python
preprocess_audio(src_path, dest_path)
```

```python
n = len(os.listdir(agent_audio_path))

for (i, audio_name) in enumerate(os.listdir(agent_audio_path)):
  src_path = os.path.join(agent_audio_path, audio_name)
  dest_path = os.path.join(agent_pp_audio_path, audio_name)
  preprocess_audio(src_path, dest_path)

  if i % 10 == 0:
    print(f"Done: {i}/{n}")
```

```python
n = len(os.listdir(caller_audio_path))

for (i, audio_name) in enumerate(os.listdir(caller_audio_path)):
  src_path = os.path.join(caller_audio_path, audio_name)
  dest_path = os.path.join(caller_pp_audio_path, audio_name)
  preprocess_audio(src_path, dest_path)

  if i % 50 == 0:
    print(f"Done: {i}/{n}")
```
